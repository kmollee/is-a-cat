```python
path = untar_data(URLs.PETS)/'images'        # line 2
```

```python
def is_cat(x): return x[0].isupper()
dls = ImageDataLoaders.from_name_func(        # line 4
    path,
    get_image_files(path),
    valid_pct=0.2,
    seed=42,
    label_func=is_cat,
    item_tfms=Resize(224))
```

```python
learn = vision_learner(dls, resnet34, metrics=error_rate)        # line 5
learn.fine_tune(1)                                            # line 6
```

```python
if iskaggle:
    learn.export("/kaggle/working/model.pkl")
else:
      learn.export("./model.pkl")
#doc(learn.export)
```
